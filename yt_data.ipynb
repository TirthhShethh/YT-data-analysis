```python
import pandas as pd
import plotly.express as px

# 1. Convert SUBSCRIBERS ("274M", "21.12K") TO numeric
def to_number(x):
    x = str(x).strip()
    if x.endswith('K'):
        return float(x[:-1]) * 1_000
    if x.endswith('M'):
        return float(x[:-1]) * 1_000_000
    if x.endswith('B'):
        return float(x[:-1]) * 1_000_000_000
    return float(x)

data['SUBS_NUM'] = data['SUBSCRIBERS'].apply(to_number)

#2. Aggregate by category
cat = (
    data
    .groupby('CATEGORY', as_index=False)['SUBS_NUM']
    .sum()
    .sort_values('SUBS_NUM', ascending=False)
)

#3. Keep top 5, group the rest as "Others"
TOP_N = 5
top = cat.head(TOP_N)
others_sum = cat['SUBS_NUM'].iloc[TOP_N:].sum()

cat_final = top.copy()
if others_sum > 0:
    cat_final = pd.concat([
        top,
        pd.DataFrame({'CATEGORY': ['Others'], 'SUBS_NUM': [others_sum]})
    ], ignore_index=True)

#4A. Pie chart: share of subscribers by category
fig_pie = px.pie(
    cat_final,
    values='SUBS_NUM',
    names='CATEGORY',
    title='Share of Total Subscribers by YouTube Category',
    hole=0.4
)
fig_pie.update_traces(textposition='inside', textinfo='percent+label')
fig_pie.show()

#4B. Bar chart: total subscribers by category (top 10)
top10 = cat.head(10)

fig_bar = px.bar(
    top10,
    x='CATEGORY',
    y='SUBS_NUM',
    title='Top 10 YouTube Categories by Total Subscribers'
)
fig_bar.update_layout(xaxis_tickangle=-45)
fig_bar.show()
```
